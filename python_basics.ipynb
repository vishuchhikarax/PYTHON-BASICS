#Python Basics

1.  What is Python, and why is it popular?
 - Python is a programming language that's easy to read and write. People like it because it's simple to
 learn, works for many things ( like websites, games, data, and AI ), and has tons of tools already built in.
2.   What is an interpreter in Python?
  - An interpreter is like a translator that reads your Python code one line at a time and runs it right away.
3.   What are pre-defined keywords in Python?
  - Keywords are special words in Python like if, else, while, and for. These words already have a job,
  so you can't use them for other things like naming a variable.
4.   Can keywords be used as variable names?
  - Nope! You can't use keywords as variable names because Python needs them to understand your code.
5.   What is mutability in Python?
- Mutability means whether something can be changed or not.
- If it can change (like a list), it's mutable.
- If it can't change (like a tuple), it's immutable.
6.   Why are lists mutable, but tuples are not?
- Lists are made to be flexible, so you can add or remove stuff. Tuples are meant to stay the same
after you make them - kind of like a locked box.
7.   What's the difference between == and is in Python?
- == checks if two things look the same (same value).
- is checks if they are actually the same object in memory.
8.   What are logical operators in Python?
Logical operators help you combine conditions:
- and: Both things must be true
- or: At least one must be true
- not: Flips true to false, and false to true
9.   What is type casting in Python?
- Type casting is when you change a value from one type to another, like turning a number into a
string, or a string into a number.
10.   What's the difference between implicit and explicit type casting?
- Implicit casting: Python changes the type for you automatically.
- Explicit casting: You tell Python to change it using something like int() or str().
11.   What's the point of conditional statements in Python?
- They let your code make decisions, like "if this happens, do that."
12.   How does elif work?
- elif means "else if." It lets you check more than one condition. If the first if isn't true, Python checks
the elif, and so on.
13.   What's the difference between for and while loops?
- Use a for loop when you know how many times to repeat something.
- Use a while loop when you don't know how many times - it keeps going while a condition is true.
14.   When is a while loop better than a for loop?
- If you're asking a user to type something (like a password), and you don't know how many tries
they'll take, a while loop is better.

```python
#Q1 48 Write a Python program to print "Hello, World!"?
print('hello,world!')

#Q2 Write a Python program that displays your name and age.
name=input('Etner your name: ')
age=input('Enter your age: ')
print(name,age)

#Q3  Write code to print all the pre-defined keywords in Python using the keyword
import keyword
print("Python Keywords:")
print(keyword.kwlist)

#Q4 Write a program that checks if a given word is a Python keyword
import keyword
word = input("Enter a word: ")
if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is not a Python keyword.")

#Q5 Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
#for each.
my_list = [10, 20, 30]
print("Original list:", my_list)
my_list[0] = 100
print("Modified list:", my_list)
my_tuple = (10, 20, 30)
print("Original tuple:", my_tuple)
try:
    my_tuple[0] = 100
except TypeError as e:
    print("Error trying to modify tuple:", e)

#Q6 Write a function to demonstrate the behavior of mutable and immutable arguments
def demonstrate_mutable_immutable():

    def modify_list(lst):
        lst.append(100)


    def modify_int(n):
        n += 10


    mutable_arg = [1, 2, 3]
    immutable_arg = 5

    print("Before modification:")
    print("Mutable argument (list):", mutable_arg)
    print("Immutable argument (int):", immutable_arg)


    modify_list(mutable_arg)
    modify_int(immutable_arg)

    print("\nAfter modification:")
    print("Mutable argument (list):", mutable_arg)
    print("Immutable argument (int):", immutable_arg)


demonstrate_mutable_immutable()


#Q7  Write a program that performs basic arithmetic operations on two user-input numbers

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))


addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Cannot divide by zero"


print("\nResults:")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")


#Q8 Write a program to demonstrate the use of logical operators.

a = True
b = False

print("a =", a)
print("b =", b)

# Logical AND
print("a and b:", a and b)

# Logical OR
print("a or b:", a or b)

# Logical NOT
print("not a:", not a)
print("not b:", not b)


#Q9 Write a Python program to convert user input from string to integer, float, and boolean types


user_input = input("Enter a value: ")


try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to integer.")


try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to float.")


bool_value = bool(user_input)
print("Boolean:", bool_value)


\

#Q10 Write code to demonstrate type casting with list elements.

str_list = ["10", "20", "30", "40"]

int_list = [int(x) for x in str_list]

print("Original list (strings):", str_list)
print("Converted list (integers):", int_list)


#11  Write a program that checks if a number is positive, negative, or zero.
num = float(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")


#12 Write a for loop to print numbers from 1 to 10.
for i in range(1, 11):
    print(i)


#13 Write a Python program to find the sum of all even numbers between 1 and 50
sum_even = 0

for num in range(1, 51):
    if num % 2 == 0:
        sum_even += num

print("The sum of all even numbers from 1 to 50 is:", sum_even)


#14  Write a program to reverse a string using a while loop

text = input("Enter a string: ")

reversed_text = ""
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)


#15 Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1

    while i <= num:
        factorial *= i
        i += 1

    print(f"The factorial of {num} is {factorial}")

```
